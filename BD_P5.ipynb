```python
# Dataset teks sederhana
dataset = [
    "apel merah manis",
    "pisang kuning manis",
    "apel hijau segar",
    "jeruk manis segar"
]
```

```python
def map_function(text):
    """
    Fungsi Map: mengubah teks menjadi pasangan (kata, 1)
    Input: string kalimat
    Output: list of tuples [(kata, 1), (kata, 1), ...]
    """
    words = text.split()  # Memisahkan kata-kata
    result = []

    for word in words:
        word = word.lower().strip()  # Ubah ke huruf kecil dan hapus spasi
        result.append((word, 1))     # Buat pasangan (kata, 1)

    return result

# Test fungsi map
print("Contoh Map Function:")
sample_text = "apel merah manis"
mapped = map_function(sample_text)
print(f"Input: '{sample_text}'")
print(f"Output: {mapped}")
```

```python
def reduce_function(mapped_data):
    """
    Fungsi Reduce: menggabungkan pasangan dengan key yang sama
    Input: list of tuples [(kata, count), ...]
    Output: dictionary {kata: total_count}
    """
    word_count = {}

    for word, count in mapped_data:
        if word in word_count:
            word_count[word] += count  # Tambahkan ke count yang sudah ada
        else:
            word_count[word] = count   # Buat entry baru

    return word_count

# Test fungsi reduce
print("\nContoh Reduce Function:")
sample_mapped = [('apel', 1), ('merah', 1), ('manis', 1), ('apel', 1)]
reduced = reduce_function(sample_mapped)
print(f"Input: {sample_mapped}")
print(f"Output: {reduced}")
```

```python
def mapreduce_word_count(dataset):
    """
    Implementasi lengkap MapReduce untuk menghitung kata
    """
    print("=== PROSES MAPREDUCE ===\n")

    # FASE MAP
    print("FASE MAP:")
    all_mapped_data = []

    for i, text in enumerate(dataset):
        mapped = map_function(text)
        all_mapped_data.extend(mapped)
        print(f"Kalimat {i+1}: '{text}' -> {mapped}")

    print(f"\nHasil gabungan Map: {all_mapped_data}")

    # FASE REDUCE
    print("\nFASE REDUCE:")
    final_result = reduce_function(all_mapped_data)

    print("Hasil akhir - Jumlah setiap kata:")
    for word, count in sorted(final_result.items()):
        print(f"  '{word}': {count} kali")

    return final_result

# Jalankan MapReduce
dataset = [
    "apel merah manis",
    "pisang kuning manis",
    "apel hijau segar",
    "jeruk manis segar"
]

result = mapreduce_word_count(dataset)
```

```python
# ========================================
# MAPREDUCE - MENGHITUNG KATA
# ========================================

# Dataset
dataset = [
    "apel merah manis",
    "pisang kuning manis",
    "apel hijau segar",
    "jeruk manis segar"
]

# Fungsi Map
def map_function(text):
    words = text.split()
    result = []
    for word in words:
        word = word.lower().strip()
        result.append((word, 1))
    return result

# Fungsi Reduce
def reduce_function(mapped_data):
    word_count = {}
    for word, count in mapped_data:
        if word in word_count:
            word_count[word] += count
        else:
            word_count[word] = count
    return word_count

# Implementasi MapReduce
def mapreduce_word_count(dataset):
    print("=== PROSES MAPREDUCE ===\n")

    # FASE MAP
    print("FASE MAP:")
    all_mapped_data = []

    for i, text in enumerate(dataset):
        mapped = map_function(text)
        all_mapped_data.extend(mapped)
        print(f"Kalimat {i+1}: '{text}' -> {mapped}")

    print(f"\nHasil gabungan Map: {all_mapped_data}")

    # FASE REDUCE
    print("\nFASE REDUCE:")
    final_result = reduce_function(all_mapped_data)

    print("Hasil akhir - Jumlah setiap kata:")
    for word, count in sorted(final_result.items()):
        print(f"  '{word}': {count} kali")

    return final_result

# Jalankan program
print("Dataset awal:")
for i, text in enumerate(dataset):
    print(f"{i+1}. {text}")

print("\n" + "="*50)
result = mapreduce_word_count(dataset)

print("\n" + "="*50)
print("RINGKASAN HASIL:")
print(f"Total kata unik: {len(result)}")
print(f"Total kata keseluruhan: {sum(result.values())}")
```
